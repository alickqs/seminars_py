```python
def age_group(age):
    if age < 12:
        return "kid"
    if 18 > age >= 12:
        return "teen"
    if 25>age >=18:
        return "young adult"
    if 40>age>=25:
        return "adult"
    else:
        return "old!"

df["Group"] = df.Age.apply(age_group)
```
